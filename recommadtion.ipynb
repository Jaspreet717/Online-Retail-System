#Data Cleaning

```python
# Data Cleaning
df.dropna(subset=['CustomerID'], inplace=True)  # Drop rows with missing CustomerID
df['Revenue'] = df['Quantity'] * df['UnitPrice']  # Calculate Revenue for each transaction
```

#Total Revenue Calculation

```python
# 1. Calculate Total Revenue
total_revenue = df['Revenue'].sum()
print(f"Total Revenue: £{total_revenue:.2f}")
```

#  Top-Selling Products

```python
# 2. Top-Selling Products
top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Selling Products:")
print(top_products)
```

#  Sales Trend Over Time

```python
# 3. Sales Trend Over Time
df.set_index('InvoiceDate', inplace=True)
sales_trend = df['Revenue'].resample('D').sum()
```

# Plotting the sales trend

```python
# Plotting the sales trend
import matplotlib.pyplot as plt # Import the plotting library



# Plotting the sales trend
plt.figure(figsize=(10, 6))
plt.plot(sales_trend, marker='o')
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Revenue (£)')
plt.grid(True)
plt.show()
```

# Customer Segmentation Based on Purchase Behavior

```python
# Load the dataset
data = pd.read_excel(file_path, nrows=1000)  # Loading a sample of 1000 rows

# Data Cleaning
data.dropna(subset=['CustomerID'], inplace=True)
data['Revenue'] = data['Quantity'] * data['UnitPrice']  # Calculate Revenue

# Group by CustomerID and calculate total revenue per customer
customer_revenue = data.groupby('CustomerID')['Revenue'].sum()

# Define spending thresholds
bins = [0, 100, 500, 1000, 5000]
labels = ['Low', 'Medium', 'High', 'Very High']

# Segment customers based on their total spending
customer_segments = pd.cut(customer_revenue, bins=bins, labels=labels)

# Display the customer segments
print(customer_segments.value_counts())
```

# Product Popularity by Country

```python
# Load the dataset
data = pd.read_excel(file_path, nrows=1000)

# Data Cleaning
data.dropna(subset=['CustomerID'], inplace=True)

# Group by Country and Description to count product popularity
country_product_popularity = data.groupby(['Country', 'Description'])['Quantity'].sum()

# Identify the most popular product in each country
most_popular_products = country_product_popularity.groupby(level=0).idxmax()

# Display the most popular product in each country
print("\nMost Popular Product in Each Country:")
print(most_popular_products)
```

#Identify Customers with High Returns

```python
# Load the dataset
data = pd.read_excel(file_path, nrows=1000)

# Data Cleaning
data.dropna(subset=['CustomerID'], inplace=True)

# Filter for returns (negative quantities)
returns = data[data['Quantity'] < 0]

# Group by CustomerID and count the number of returns
customer_returns = returns.groupby('CustomerID')['Quantity'].count()

# Display customers with the highest number of returns
print("\nCustomers with the Most Returns:")
print(customer_returns.sort_values(ascending=False).head(10))
```

#Sales Contribution by Country

```python
# Load the dataset
data = pd.read_excel(file_path, nrows=1000)

# Data Cleaning
data.dropna(subset=['CustomerID'], inplace=True)
data['Revenue'] = data['Quantity'] * data['UnitPrice']  # Calculate Revenue

# Group by Country to calculate total revenue per country
country_revenue = data.groupby('Country')['Revenue'].sum()

# Calculate the percentage contribution of each country
country_revenue_percentage = (country_revenue / country_revenue.sum()) * 100

# Display the revenue contribution of each country
print("\nRevenue Contribution by Country (%):")
print(country_revenue_percentage.sort_values(ascending=False))
```

#Monthly Sales Analysis

```python
# Load the dataset
data = pd.read_excel(file_path, nrows=1000)

# Data Cleaning
data.dropna(subset=['CustomerID'], inplace=True)
data['Revenue'] = data['Quantity'] * data['UnitPrice']  # Calculate Revenue

# Set InvoiceDate as the index
data.set_index('InvoiceDate', inplace=True)

# Group by month and calculate total revenue
monthly_sales = data['Revenue'].resample('M').sum()

# Display monthly sales
print("\nMonthly Sales Revenue:")
print(monthly_sales)

# Plotting the monthly sales trend
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales, marker='o')
plt.title('Monthly Sales Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue (£)')
plt.grid(True)
plt.show()
```
